```python
import torch
import os
import numpy as np
from pyannote.audio import Inference, Model

# 1. SETUP
device = torch.device("cuda")
checkpoint_path = "lightning_logs/version_1/checkpoints/hindi_best.ckpt"
trained_model = Model.from_pretrained(checkpoint_path)
trained_model.to(device)

# We use the default inference but we will handle the reshaping manually
inference = Inference(trained_model, device=device)

def get_truth_matrix(path, total_frames, speakers=3):
    truth = np.zeros((total_frames, speakers))
    # We'll calculate a 'pseudo-step' to map RTTM time to frame index
    # based on the total frames the model actually produced
    with open(path, "r") as f:
        lines = f.readlines()
        if not lines: return truth

        # Find max duration in RTTM to scale
        max_time = 0
        for line in lines:
            p = line.strip().split()
            if len(p) >= 5:
                max_time = max(max_time, float(p[3]) + float(p[4]))

        scale = total_frames / max_time if max_time > 0 else 1

        for line in lines:
            parts = line.strip().split()
            if len(parts) >= 8:
                start_f = int(float(parts[3]) * scale)
                end_f = int((float(parts[3]) + float(parts[4])) * scale)
                label = parts[7]
                idx = int(''.join(filter(str.isdigit, label))) % speakers if any(c.isdigit() for c in label) else 0
                truth[start_f:min(end_f, total_frames), idx] = 1
    return truth

# 2. EVALUATION
TEST_AUDIO_DIR = "/content/dataset/test/audio"
TEST_RTTM_DIR = "/content/dataset/test/rttm"

all_accuracies = []
print("🧪 Flattening 3D Chunks and Calculating Accuracy...")

audio_files = sorted([f for f in os.listdir(TEST_AUDIO_DIR) if f.endswith(".wav")])

for filename in audio_files:
    uri = os.path.splitext(filename)[0]
    rttm_path = os.path.join(TEST_RTTM_DIR, f"{uri}.rttm")

    if os.path.exists(rttm_path):
        try:
            output = inference(os.path.join(TEST_AUDIO_DIR, filename))

            # RESHAPE FIX: Convert (Windows, Frames, Speakers) -> (Windows * Frames, Speakers)
            # This turns the 3D block into a 2D timeline
            raw_data = np.array(output.data)
            if len(raw_data.shape) == 3:
                num_windows, frames_per_w, num_speakers = raw_data.shape
                predictions_2d = raw_data.reshape(-1, num_speakers)
            else:
                predictions_2d = raw_data

            preds_binary = (predictions_2d > 0.5).astype(int)
            total_frames = preds_binary.shape[0]

            # Get matching truth matrix
            truth_2d = get_truth_matrix(rttm_path, total_frames)

            # Final Calculation
            acc = (preds_binary == truth_2d).mean()
            all_accuracies.append(acc)
            print(f"✅ {uri}: {acc*100:.2f}% Match")

        except Exception as e:
            print(f"❌ Error on {uri}: {str(e)}")

if all_accuracies:
    final_score = np.mean(all_accuracies) * 100
    print("\n" + "="*40)
    print(f"🎯 TRUE FRAME-LEVEL ACCURACY: {final_score:.2f}%")
    print("="*40)
```

```python
import torch
import os
import numpy as np
from pydub import AudioSegment
from pyannote.audio import Inference, Model
from pyannote.audio.utils.signal import Binarize
from pyannote.core import Annotation, Segment
import matplotlib.pyplot as plt

# 1. AUDIO PREPROCESSING
def prepare_audio(input_path):
    target_path = "processed_audio.wav"
    audio = AudioSegment.from_file(input_path)
    audio = audio.set_frame_rate(16000).set_channels(1)
    audio.export(target_path, format="wav")
    return target_path

# 2. SETUP MODEL
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
checkpoint_path = "lightning_logs/version_1/checkpoints/hindi_best.ckpt"

model = Model.from_pretrained(checkpoint_path)
model.to(device)

# Using 'sliding' window but we will aggregate manually
inference = Inference(model, device=device, window="sliding", duration=5.0, step=0.5)

# 3. RUN ON YOUR FILE
USER_FILE = "/content/A28.wav"

if os.path.exists(USER_FILE):
    print(f"🔄 Converting {USER_FILE}...")
    clean_wav = prepare_audio(USER_FILE)

    print(f"🎙️ Analyzing Hindi audio...")
    outputs = inference(clean_wav)

    # --- THE FIX: MANUAL AGGREGATION ---
    # Convert SlidingWindowFeature to a 2D array (Time, Speakers)
    raw_data = np.array(outputs.data)
    if len(raw_data.shape) == 3:
        # Collapse the windows into a continuous timeline
        # We take the max activation across overlapping windows
        data_2d = raw_data.reshape(-1, raw_data.shape[-1])
    else:
        data_2d = raw_data

    # Apply the 0.32 Threshold manually since the binarizer class is picky
    print("\n--- Generating Timeline ---")
    fig, ax = plt.subplots(figsize=(12, 4))
    colors = ['#4285F4', '#EA4335', '#FBBC05', '#34A853']

    # Calculate time per frame
    # Total duration / total frames
    duration = outputs.sliding_window.duration + outputs.sliding_window.step * (len(raw_data)-1)
    time_axis = np.linspace(0, duration, len(data_2d))

    # Plot each speaker's activity
    for i in range(data_2d.shape[1]):
        activity = data_2d[:, i] > 0.32  # Your optimized threshold
        if np.any(activity):
            # Find continuous segments for the Gantt chart
            # (Simplifying: plotting the probability line for clarity)
            ax.plot(time_axis, data_2d[:, i] + (i * 1.2), label=f"Speaker {i+1}", color=colors[i % len(colors)])
            ax.fill_between(time_axis, i * 1.2, data_2d[:, i] + (i * 1.2), alpha=0.3, color=colors[i % len(colors)])

    ax.set_yticks([i * 1.2 + 0.5 for i in range(data_2d.shape[1])])
    ax.set_yticklabels([f"Speaker {i+1}" for i in range(data_2d.shape[1])])
    ax.set_xlabel("Time (seconds)")
    ax.set_ylabel("Activity Probability + Offset")
    plt.title(f"Hindi Diarization Result: {os.path.basename(USER_FILE)}")
    plt.grid(axis='x', alpha=0.3)
    plt.legend()
    plt.show()

else:
    print(f"❌ File not found: {USER_FILE}")
```

```python
import torch
import os
import glob
import numpy as np
from pyannote.audio import Model, Inference
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

# 1. LOAD MODELS (Assuming audio is already processed from previous runs)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

checkpoints = glob.glob("lightning_logs/version_*/checkpoints/*.ckpt")
best_model_path = max(checkpoints, key=os.path.getctime)
print(f"📦 Using Model: {best_model_path}")

# Load embedding model with 5s window
embedding_model = Inference("pyannote/embedding",
                            window="sliding",
                            duration=5.0,
                            step=0.5,
                            use_auth_token=True).to(device)

# 2. PROCESS
USER_FILE = "/content/A1.wav"

if os.path.exists(USER_FILE):
    # Get Voiceprints
    embeddings = embedding_model(USER_FILE)
    emb_data = embeddings.data

    # Normalize
    scaler = StandardScaler()
    emb_scaled = scaler.fit_transform(emb_data)

    # 3. THE CALIBRATED CLUSTERER
    # We are moving to 0.42.
    # Logic: If you get too many speakers, INCREASE this number (try 0.45).
    # If you get only 1 speaker, DECREASE this number (try 0.38).
    clusterer = AgglomerativeClustering(
        n_clusters=None,
        distance_threshold=0.42,
        linkage='average' # 'average' is more stable for diarization
    )

    labels = clusterer.fit_predict(emb_scaled)

    # 4. CLEAN OUTPUT LOGIC
    print(f"\n✨ DIARIZATION LOG: {os.path.basename(USER_FILE)}")
    print("-" * 45)

    current_spk = -1
    start_time = 0
    step = embeddings.sliding_window.step

    for i, label in enumerate(labels):
        timestamp = i * step
        if label != current_spk:
            if current_spk != -1:
                # Only print if the speaker spoke for more than 0.5 seconds
                if (timestamp - start_time) > 0.5:
                    print(f"Speaker {current_spk + 1}: {start_time:06.2f}s ---> {timestamp:06.2f}s")
            current_spk = label
            start_time = timestamp

    # Print the last segment
    print(f"Speaker {current_spk + 1}: {start_time:06.2f}s ---> {(len(labels)*step):06.2f}s")

    print("-" * 45)
    print(f"✅ FINAL COUNT: {len(np.unique(labels))} speakers detected.")
```
